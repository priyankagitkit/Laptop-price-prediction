```python
X = df1.drop(["Price","Strcap"],axis = 1) # By looking the data we found that HDD and SSD have the same thing as Strcap so removing the Strcap column

y =df1['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from xgboost import XGBRegressor
```

```python
models = []

models.append(('RF', RandomForestRegressor()))
models.append(('GB', GradientBoostingRegressor()))
models.append(('DT', DecisionTreeRegressor()))
models.append(('SVM', SVR()))
models.append(('XGboost',XGBRegressor()))
models
```

```python
# Assuming 'models' is a list of (name, model) tuples
results = []
names = []

for name, model in models:
    
    # Assuming model is a RandomForestRegressor instance
    
    model.fit(X_train, y_train)  # Fit the model with your training data
    
    # Now you can make predictions
    
    y_pred = model.predict(X_test)
    
    # Use r2_score with predicted and true values
    
    r2_result = r2_score(y_test, y_pred)
    mae_result = mean_absolute_error(y_test, y_pred)
    
    results.append(r2_result)
    names.append(name)
    
    msg = "%s: R-squared = %f, MEAN_ABSOLUTE_ERROR = %f" % (name, r2_result, mae_result)
    print(msg)
```
