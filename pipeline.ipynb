```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_preprocess.csv')
df = df.iloc[:,1:]
df1 = df.copy()
df1
```

```python
df1['Gpu_brand'].value_counts()
```

```python
X = df1.iloc[:,:-1]
y = df1['Price']
```

```python
X.head(1)
```

```python
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
dm = DataFrameMapper([(["Company",LabelEncoder()]),
                      (["TypeName",LabelEncoder()]),
                      (["Cpu_brand",LabelEncoder()]),
                      (["Gpu_brand",LabelEncoder()]),
                      (["Os",LabelEncoder()])])
```

```python
dm
```

```python
ml = []
ml.append(['mapper',dm])
ml.append(['SC',StandardScaler()])
ml.append(['RF',RandomForestRegressor()])
```

```python
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

```python
model = Pipeline(ml)
model.fit(X,y)
```
